```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_stle="whitegrid"
```

```python
Books_data=pd.read_csv("books.csv")
Books_data.head(5)
```

```python
Books_data["rating"]=Books_data[["ratings_1","ratings_2","ratings_3","ratings_4","ratings_5"]].mean(axis=1)
```

```python
Books_data.drop(["original_title","id","best_book_id","isbn13","small_image_url","ratings_1","ratings_2","ratings_3","ratings_4","ratings_5"],axis=1,inplace=True)
Books_data.head(5)
```

```python
Books_data["total_ratings_count"]=Books_data["ratings_count"]+Books_data["work_ratings_count"]+Books_data["work_text_reviews_count"]+Books_data["rating"]
Books_data.drop(["ratings_count","work_ratings_count","work_text_reviews_count","rating"],axis=1,inplace=True)
Books_data.head(5)
```

```python
Books_data.drop(["work_id","average_rating"],axis=1,inplace=True)
```

```python
tags=pd.read_csv("tags.csv")
tags.head(5)
```

```python
book_tag= pd.read_csv("book_tags.csv")
book_tag.head(5)
```

```python
df1=book_tag.merge(tags)
```

```python
df1.head(5)
```

```python
df2 = df1.groupby(['goodreads_book_id']).mean().reset_index()
df2['book_mean_count'] = df2['count'].round()

df2.drop(['count',"tag_id"],axis=1,inplace=True)
```

```python
df2.head(5)
```

```python
ratings=pd.read_csv("ratings.csv")
ratings.head(5)
```

```python
to_read=pd.read_csv("to_read.csv")
to_read.head(5)
```

```python
df2=pd.concat([ratings,to_read])
df2.head(5)
```

# To Grab Best Rated books

```python
df2.groupby("book_id")["rating"].mean().sort_values(ascending=False).head()
```

# To Get Mean Ratings Per User

```python

MRPU = df2.groupby(['book_id']).mean().reset_index()
MRPU['book_mean_rating'] = MRPU['rating'].round(2)

MRPU.drop(['user_id','rating'],axis=1,inplace=True)
```

```python
MRPU.head(5)
```

```python
MRPU.info()
```

```python
Books_data.info()
```

```python
data=Books_data.merge(MRPU,on="book_id")
```

```python
data.head(5)
```

```python
data["book_mean_rating"].hist(bins=50)
```

# Sorting on the basis total ratings count

```python
data.sort_values("total_ratings_count",ascending=False).head(5)
```

# To get top 20 rated books

```python
top_20_rated=data.groupby("title")["book_mean_rating"].sum().reset_index().sort_values(by="book_mean_rating",ascending=False)
```

```python
top_20_rated.head(20)
```

# To get books having similar ratings as "Hard Times"

```python
similar_to_hardtimes_in_rating=top_20_rated[(top_20_rated['book_mean_rating']>=4.5)]
```
